###**Soal**
3. Buat program untuk melakukan perbandingan deteksi tepi dari beberapa operator menggunakan gambar yang kalian dapat dari handphone masing-masing (sehingga masing-masing mahasiswa gambar berbeda).

###**1. Import Library**

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, filters, feature
from skimage.filters.rank import gradient
from skimage.morphology import disk
```

###**2. Membaca gambar dari file**

```python
image = io.imread("pic.jpeg")
```

###**3. Konversi ke Grayscale**

```python
gray = color.rgb2gray(image)
```

###**4. Terapkan berbagai operator deteksi tepi**

```python
sobel = filters.sobel(gray)
roberts = filters.roberts(gray)
prewitt = filters.prewitt(gray)
kirsch = gradient((gray * 255).astype(np.uint8), disk(1))
canny = feature.canny(gray, sigma=1)
```

###**5. Visualisasi hasil perbandingan**

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
ax = axes.ravel()

ax[0].imshow(image)
ax[0].set_title("Gambar Asli")
ax[0].axis("off")

ax[1].imshow(sobel, cmap='gray')
ax[1].set_title("Sobel")
ax[1].axis("off")

ax[2].imshow(roberts, cmap='gray')
ax[2].set_title("Roberts")
ax[2].axis("off")

ax[3].imshow(prewitt, cmap='gray')
ax[3].set_title("Prewitt")
ax[3].axis("off")

ax[4].imshow(kirsch, cmap='gray')
ax[4].set_title("Kirsch")
ax[4].axis("off")

ax[5].imshow(canny, cmap='gray')
ax[5].set_title("Canny")
ax[5].axis("off")

plt.tight_layout()
plt.show()
```
